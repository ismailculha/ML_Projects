```python
df[df["Aircraft Carrier Fleet Strength value"].isnull()]
```

```python
# Normally Geopraphic values are very important for military strength, bu in this notebook we don't use them. So I drop theöm
df = df.drop(columns=["Total Square Land Area value","Total Coastline Coverage value","Total Waterway Coverage value","Total Border Coverage value"])
```

# Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler, RobustScaler,MinMaxScaler
from sklearn.cluster import KMeans

X = df.drop(columns=["Country"])
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

# Modelling

```python
K_means_model = KMeans(n_clusters=5, random_state=42)
K_means_model.fit_predict(X)
K_means_model.labels_
```

## Hopkins Test 
#### Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express no clusterability, a score tending to 1 express an uniform distrubution so clustering can't be useful and  0 express  not an uniform distrubution hence clusterering can be useful.

```python
from pyclustertend import hopkins
hopkins(X, X.shape[0])

#Clustering tendency is good enough 
```

## Clustering with K-means

### Elbow Method 
### We use this method to find optimal k value by looking up to break point of graph. 

```python
ssd = [] #sum squared distance
K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(X)
    ssd.append(model.inertia_)
    
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error (ssd)") 
plt.title("elbow method")
```

```python
from yellowbrick.cluster import KElbowVisualizer

model_ = KMeans(random_state=42)
visualizer = KElbowVisualizer(model_, k=(2,9))

visualizer.fit(X)        # Fit the data to the visualizer
visualizer.show();
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(X, K_means_model.labels_)
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
    
from sklearn.cluster import KMeans
from yellowbrick.cluster import SilhouetteVisualizer

model_k_mean = KMeans(4, random_state=42)
visualizer = SilhouetteVisualizer(model_k_mean)
visualizer.fit(X)    # Fit the data to the visualizer
visualizer.poof();
```

## Hierarchical Clustering

### Model Building and label visualisation

```python
from scipy.cluster.hierarchy import dendrogram, linkage

hc_ward = linkage(y = X, method = "ward")
hc_average = linkage(y = X, method = "average")

plt.figure(figsize=(16,20))

plt.subplot(211)
dendrogram(hc_ward, leaf_font_size = 10);

plt.subplot(212)
dendrogram(hc_ward, leaf_font_size = 10);
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

K = range(2,10)

for k in K:
    model = AgglomerativeClustering(n_clusters = k,linkage='ward')
    model.fit_predict(X)
    print(f'Silhouette Score for {k} clusters: {silhouette_score(X, model.labels_)}')
```

### Evaluate 

```python
Hierarchial_model = AgglomerativeClustering(n_clusters=4, affinity = "euclidean", linkage = "ward")
clusters = Hierarchial_model.fit_predict(X)
clusters
```

```python
df_x = pd.DataFrame(X, columns=df.iloc[:,:-2].columns)
df_x
```

```python
df["Cluster_Hierarchial"] = Hierarchial_model.labels_
df_x["Cluster"] = Hierarchial_model.labels_
df["Cluster_Hierarchial"].value_counts()
```
